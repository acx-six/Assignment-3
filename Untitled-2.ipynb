**Correlation Matrix**

```python
# Generate and visualize the correlation matrix with a different heatmap style
def heatmap(X):
    corr = X.corr().round(2)  # Compute the correlation matrix
    plt.figure(figsize=(8, 8))  # Increase the figure size for clarity
    sns.heatmap(corr, cmap='magma', vmin=-1, vmax=1, center=0, 
                square=True, linewidths=0.8, cbar_kws={"shrink": .75}, annot=True, annot_kws={"fontsize": 9})
    plt.title("Feature Correlation Matrix", fontsize=14)  # Add a title to the heatmap
    plt.show()

heatmap(X)
```
